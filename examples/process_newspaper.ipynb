```python
import urllib.request
import bs4
```

Read HTML article

```python
response = urllib.request.urlopen("http://montrealgazette.com/travel/checking-in-eastern-townships-farmbb-dishes-up-barn-to-fork-cuisine")
body = response.read()
```

Start producing the soup

```python
soup = bs4.BeautifulSoup(body.decode("utf-8"),"html.parser")
```

```python
article = soup.find(itemprop="articleBody")
type(article)
```

```python
paragraphs = article.find_all("p")
```

```python
num = len(paragraphs)
num
```
